```python
import pandas as pd
import numpy as np
from lxml import objectify
import xml.etree.cElementTree as et
```

```python
xml = et.parse('grid.summ.out')
dfcols = ['time','loaded','inserted','running','waiting','ended','meanWaitingTime','meanTravelTime','duration']
df_xml = pd.DataFrame(columns=dfcols)

for node in xml.getroot():
    time = node.attrib.get('time')
    loaded = node.attrib.get('loaded')
    inserted = node.attrib.get('inserted')
    running = node.attrib.get('running')
    waiting = node.attrib.get('waiting')
    ended = node.attrib.get('ended')
    meanWaitingTime = node.attrib.get('meanWaitingTime')
    meanTravelTime = node.attrib.get('meanTravelTime')
    duration = node.attrib.get('duration')
    
    df_xml = df_xml.append(pd.Series([time,loaded,inserted,running,
                                      waiting,ended,meanWaitingTime,
                                      meanTravelTime,duration],
                                     index=dfcols), ignore_index=True)
    
    
```
